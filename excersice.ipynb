```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("white")
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
exp = pd.read_csv(url)

exp.head(10)
```

```python
del exp['Unnamed: 0']
exp.head(10)
```

```python
ttbill = sns.displot(exp.total_bill);
ttbill.set(xlabel = 'value' , ylabel = 'frequency' , title = "Total Bill")
sns.despine()
```

```python
sns.jointplot(x = "total_bill" , y = "tip" , data = exp)
```

```python
sns.pairplot(exp)
```

```python
sns.stripplot(x = "day" , y = "total_bill" , data = exp , jitter = True)
```

```python
sns.stripplot(x = "tip" , y = "day" , hue = "sex" , data = exp , jitter = True)
```

```python
sns.boxplot(x = "day" , y = "total_bill" , hue = "time" , data = exp)
```

```python
sns.set(style = "ticks")

h = sns.FacetGrid(exp , col = "time")
h.map(plt.hist , "tip")
```

```python
h = sns.FacetGrid(exp , col = "sex", hue = "smoker")
h.map(plt.scatter , "total_bill" , "tip" , alpha = .5)

h.add_legend()
```

```python
# BONUS  :   scatterplot with tip as y axis , total_bill as x axis , differ by size.
sns.scatterplot(x = "total_bill" , y = "tip" , hue = "size" , data = exp)
```
