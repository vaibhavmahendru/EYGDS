Find all of the numbers from 1-1000 that are divisible by 8.

```python
l = []

for x in range(1,1001):
  if x % 8 == 0:
    l.append(x)

print(l)

```

 Find all of the numbers from 1-1000 that have a 6 in them.

```python
nl = [x for x in range(1,1001)]

result = [x for x in nl if "6" in str(x)]

result


```

3. Count the number of spaces in the following string. 

String:  “Practice problems to drill list comprehension in your head” 

```python
s1 = "Practice problems to drill list comprehension in your head"

res = len([x for x in s1 if x == " "])

res
```

 Remove all the vowels in the above string.

```python
res = "".join([x for x in s1 if x not in ['a' ,'e','i','o','u']])
res
```

. Find all the words in the above string that are less than 5 letters.

```python
s = s1.split(" ")
res = ([x for x in s if len(x) < 5])
res
```

. use dictionary comprehension to count the length of each word in  the sentence (refer to the above string)

```python

res = {x:len(x) for x in s}

res
```
