```python
class Node:
  
    
    def __init__(self, firstname , lastname , hobby , ph):
        self.firstname = firstname
        self.lastname = lastname
        self.hobby = hobby
        self.ph = ph
        self.next = None
  
  

class LinkedList:
  
    
    def __init__(self):
        self.head = None


    def insert_node(self, firstname , lastname , hobby , ph):
        node = Node(firstname , lastname , hobby , ph)
        node.next = self.head
        self.head = node
  
  

    def delete_duplicates(self):

        current = self.head
        prev = None
        duplicate_dict = dict()
        while current:
            if current.firstname and current.lastname and current.hobby and current.ph not in duplicate_dict:
                duplicate_dict[current.ph] = None
                prev = current
            else:
                prev.next = current.next

            current = current.next

    
    def printList(self):
        data = self.head
        while (data):
            print(data.firstname , data.lastname, data.hobby , data.ph)
            data = data.next



    def reverse(self):
        prev = None
        current = self.head
        while(current is not None):
            next = current.next
            current.next = prev
            prev = current
            current = next
        self.head = prev



    def Search(self, key):
        current = self.head
 
        index = 0
        while current:
            if current.ph == key:
                current.firstname = "Jerry" 
            current = current.next
            index = index + 1
 
        return -1

   
 
  


l = LinkedList()



l.head = Node("Alex" , "Code" , "football" , 123654)
second = Node("Bob" , "Marley" , "singing" , 123655)
third = Node("Andy" , "Grammar" , "dance" , 321554)
fourth = Node("Bob" , "Marley" , "singing" , 123655)


l.head.next = second  
second.next = third  
third.next = fourth


print("initial LL")
print("\n")
l.printList()

l.insert_node("Bob" , "Marley" , "singing" , 123655)







print("\n")
print("\n")
print("inserting in LL")
print("\n")

l.printList()

l.delete_duplicates()

print("\n")
print("removing duplicates")

print("\n")




l.printList()

l.reverse()

print("\n")

print("reversing")
print("\n")

l.printList()


print("\n")
print("searching")
print("\n")

l.Search(123654)

print("\n")

l.printList()




```
